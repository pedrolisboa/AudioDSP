```python
import os
import matplotlib.pyplot as plt
import numpy as np
from misc import *
from scipy.signal import stft, blackman, hann

datapath = './../../audio/reduzidas'
```

```python
audio_map = {audiofile:read_audio(os.path.join(datapath, audiofile)) for audiofile in os.listdir(datapath)}
```

```python
%matplotlib inline
nrows = len(audio_map)
fig, axes = plt.subplots(nrows=nrows, figsize=(15,15))
plt.subplots_adjust(hspace=0.45)
for (audio_name, (_, audio)), ax in zip(audio_map.items(), axes): 
    ax.plot(audio)
    ax.set_title(audio_name[:-4].encode(encoding='UTF-8'))
    ax.set_xlabel("Frame")
    ax.set_ylabel("Amplitude")
    
```

```python
wlen = 512
win = hann(wlen)
nfft = 1024

spec_map = {audioname: stft(audio, fs, win, wlen, wlen/2, nfft) 
            for (audioname, (fs, audio)) in audio_map.items()}
```

```python
fig, axes = plt.subplots(nrows=4,ncols=1, figsize=(15,15))
plt.subplots_adjust(hspace=0.45)
for (audio_name, (f, t, Sxx)), ax in zip(spec_map.items(), axes.flat): 
    C = sum(win)/wlen;
    power = np.absolute(Sxx)/wlen/C
#     power = power / tpsw(power)#, **tpsw_args)
    power = 20*np.log10(power + 1e-6)
#     power[power < -0.2] = 0
    plot_spec(power, f, t, audio_name, ax)
    
```
